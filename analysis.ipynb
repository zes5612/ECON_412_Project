```python
import polars as pl 
# import os

# Sample ticker
data_path = "Data/Master_Data/MMM.csv"

df = pl.read_csv(data_path)

df.head()
```

```python
df.shape
```

Compute daily returns as percent change

```python
df = df.with_columns(pl.col("Close").pct_change().fill_null(0).alias("Daily Return"))
df.head(10)
```

Combine all ticker CSVs

```python
import glob

files = glob.glob("Data/Master_Data/*.csv")
dfs = [pl.read_csv(file) for file in files]

master_df = pl.concat(dfs)

master_df = master_df.with_columns(pl.col("Date").str.strptime(pl.Date, "%Y-%m-%d"))


master_df.head(20)
```

```python
master_df.shape
```

Daily Returns per Ticker

```python
master_df = master_df.with_columns(pl.col("Close").pct_change().over("Ticker").fill_null(0).alias("Daily Return"))
master_df.head(10)
```

Aggregated Average Returns and Volatility by Sector

```python
# NOT what we need; just curious
sector_summary = master_df.group_by("Sector").agg([
    pl.col("Daily Return").mean().alias("Sector Avg Daily Return"),
    pl.col("Daily Return").std().alias("Sector Volatility")
]).drop_nulls()

sector_summary
```

```python
sector_summary.shape
```

Aggregated Average Returns and Volatility by Sub-Industry

```python
# NOT what we need; just curious
industry_summary = master_df.group_by("Sub-Industry").agg([
    pl.col("Daily Return").mean().alias("Sub-Industry Avg Daily Return"),
    pl.col("Daily Return").std().alias("Sub-Industry Volatility")
]).drop_nulls()

industry_summary
```

Extract 'year' from master_df

```python
master_df = master_df.with_columns(
    pl.col("Date").dt.year().alias("Year")  # extract year
)

master_df.head()
```

Yearly Compounded Return by Sector

```python
yearly_sector_returns = master_df.group_by(["Year", "Sector"]).agg([
    ((1 + pl.col("Daily Return")).cum_prod().last() - 1).alias("Compounded Return")
]).drop_nulls().sort(["Year", "Sector"])

yearly_sector_returns
```

Daily Arithemic Return by Sector

```python
sector_daily_returns = (
    master_df
    .group_by(["Date", "Sector"])
    .agg(
        pl.col("Daily Return").mean().alias("Sector Daily Return")
    )
)

sector_daily_returns
```

```python
sector_daily_returns = sector_daily_returns.with_columns(
    pl.col("Date").dt.year().alias("Year")
)

sector_daily_returns
```

Annualized Sector Volatility

```python
annual_sector_volatility = (
    sector_daily_returns
    .group_by(["Year", "Sector"])
    .agg(
        (pl.col("Sector Daily Return").std() * (252 ** 0.5)).alias("Annual Volatility")
    )
) # ~252 trading days

annual_sector_volatility
```
