```python
import polars as pl 
# import os

# Sample ticker
data_path = "Data/Master_Data/MMM.csv"

df = pl.read_csv(data_path)

df.head()
```

Compute daily returns as percent change

```python
df = df.with_columns(pl.col("Close").pct_change().fill_null(0).alias("Daily Return"))
df.head(10)
```

Combine all ticker CSVs

```python
import glob

files = glob.glob("Data/Master_Data/*.csv")
dfs = [pl.read_csv(file) for file in files]

master_df = pl.concat(dfs)

master_df = master_df.with_columns(pl.col("Date").str.strptime(pl.Date, "%Y-%m-%d"))


master_df.head(20)
```

```python
master_df.shape
```

Daily Returns per Ticker

```python
master_df = master_df.with_columns(pl.col("Close").pct_change().over("Ticker").fill_null(0).alias("Daily Return"))
#master_df.filter(pl.col(""))
```

Aggregated Average Returns and Volatility by Sector

Aggregated Average Returns and Volatility by Sub-Industry

```python
master_df
daily_industry_returns = master_df.group_by(["Date", "Sub-Industry"]).agg(
    [pl.col("Daily Return").mean().alias("Sub-Industry Daily Return"),
    pl.col("Volume").mean().alias("Sub-Industry Daily Volume")]
)

daily_industry_returns = daily_industry_returns.with_columns(
    pl.col("Date").dt.year().alias("Year") 
)


yearly_industry_returns = daily_industry_returns.group_by(["Year", "Sub-Industry"]).agg([
    ((1 + pl.col("Sub-Industry Daily Return")).cum_prod().last() - 1).alias("Compounded Return"),
    pl.col("Sub-Industry Daily Volume").sum().alias("Total Volume"),
    (pl.col("Sub-Industry Daily Return").std() * (252 ** 0.5)).alias("Annual Volatility")
]).drop_nulls().sort(["Year", "Sub-Industry"])


#yearly_industry_returns.write_csv("industry_compounded_yearly_returns.csv")
yearly_industry_returns
```

Extract 'year' from master_df

```python
master_df = master_df.with_columns(
    pl.col("Date").dt.year().alias("Year")  # extract year
)

master_df.head()
```

Yearly Compounded Return by Sector

```python
daily_sector_returns = master_df.group_by(["Date", "Sector"]).agg(
    pl.col("Daily Return").mean().alias("Sector Daily Return")
)
daily_sector_returns = daily_sector_returns.with_columns(
    pl.col("Date").dt.year().alias("Year") 
)
print(daily_sector_returns)

yearly_sector_returns = daily_sector_returns.group_by(["Year", "Sector"]).agg([
    ((1 + pl.col("Sector Daily Return")).cum_prod().last() - 1).alias("Compounded Return")]
).drop_nulls().sort(["Year", "Sector"])

yearly_sector_returns.with_columns
print(yearly_sector_returns)

#yearly_sector_returns.write_csv("Sector_compounded_yearly_returns.csv")
```

Daily Arithemic Return by Sector

```python
sector_daily_returns = (
    master_df
    .group_by(["Date", "Sector"])
    .agg(
        pl.col("Daily Return").mean().alias("Sector Daily Return")
    )
)

sector_daily_returns
```

```python
sector_daily_returns = sector_daily_returns.with_columns(
    pl.col("Date").dt.year().alias("Year")
)

sector_daily_returns
```

Annualized Sector Volatility

```python
annual_sector_volatility = (
    sector_daily_returns
    .group_by(["Year", "Sector"])
    .agg(
        (pl.col("Sector Daily Return").std() * (252 ** 0.5)).alias("Annual Volatility")
    )
) # ~252 trading days

annual_sector_volatility
```
